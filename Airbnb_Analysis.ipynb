Importing the dataset should come first, then come data cleansing and visualization.

**Importing libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt     # Visualization
%matplotlib inline
import seaborn as sns               # Visualization
```

```python
# Read the CSV file (Listings.csv)
listings_df = pd.read_csv('Listings.csv', encoding='ISO-8859-1')
listings_df
```

```python
# Read the CSV file (Reviews.csv)
reviews_df = pd.read_csv('Reviews.csv', encoding='ISO-8859-1')
reviews_df
```

```python
# Merge based on a (listing_id) >> common column
merged_df = pd.merge(listings_df, reviews_df, on='listing_id')
```

```python
# Print them
merged_df.head()
```

**Data Loading and Initial Exploration**

```python
# Print columns
merged_df.columns
```

```python
# Print the number of rows and columns
merged_df.shape
```

```python
# Basic information about the dataset
merged_df.info()
```

**Data Cleaning**

```python
# Duplicate check of rows in dataset
merged_df = merged_df.drop_duplicates()
merged_df.count()
```

```python
# Another way (Not necessary)!
duplicate_rows = merged_df[merged_df.duplicated()]
print("Duplicate rows: ")
print(duplicate_rows)
```

```python
# Null check, sum for calculte total Null in each column
merged_df.isnull().sum()
```

```python
# Remove missing values (NaN)
merged_df.dropna()
```

```python
# Drops all the values
merged_df.dropna(how='all')
```

```python
# Fill in with zero
merged_df.fillna(0,inplace=True)
merged_df
```

```python
# Another check for Null to ensure there is no null values
merged_df.isnull().sum()
```

```python
# Count the number of cities
counts = listings_df['city'].value_counts()

top_city = counts.reset_index()

top_city.columns = ['city', 'Counts']
top_city
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(listings_df['city'])

# Title
plt.title('Top City', fontsize=15)

# x-axis label
plt.xlabel('Counts', fontsize=14)

# y-axis label
plt.ylabel('City', fontsize=14)
```

**Observations:**

*   Paris and New York are the top two cities.  

*   Hong kong is the lowest city, with approximately 90000.

*   Bangkok and Mexico City approximately have the same counts.
